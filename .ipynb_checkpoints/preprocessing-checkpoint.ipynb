```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
```

```python
df = pd.read_csv("data/train_2008.csv")
df.head()
```

```python
# remove unnecessary features

new_df = df.copy()

for feature in df.columns.values:
    temp = new_df[feature]
    elem0 = temp[0]
    if np.count_nonzero(temp-elem0) == 0:
        new_df = new_df.drop(feature, 1)
new_df = new_df.drop('id', 1)
```

```python
new_df.head()
```

```python
print(new_df.shape)
```

```python
# base line something
x = new_df.loc[:, new_df.columns != 'target'].values
y = new_df['target'].values

print(x.shape)
print(y.shape)
```

```python
# Shuffle order of data

rand_order = np.arange(len(x))
np.random.seed(42)
np.random.shuffle(rand_order)
print(rand_order)

new_x = []
new_y = []

for i in range(len(rand_order)):
    new_x.append(x[i])
    new_y.append(y[i])

new_x = np.array(new_x)
new_y = np.array(new_y)

print(new_x.shape)
print(new_y.shape)
```

```python
np.save("data/x.npy", new_x)
np.save("data/y.npy", new_y)
```
