```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split

x = np.load("data/x.npy")
y = np.load("data/y.npy")
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.decomposition import PCA


colors = []
for i, _ in enumerate(x_train):
    if y_train[i] == 0:
        colors.append('b')
    else:
        colors.append('g')
    
    
plt.scatter(x_train[:, 0], x_train[:, 1], c=colors)
plt.show()

pca_model = PCA(n_components=2)
pca_model.fit(x_train)
embeddings = pca_model.transform(x_train)
embeddings.shape
```

```python
colors = []
for i, _ in enumerate(embeddings):
    if y_train[i] == 0:
        colors.append('b')
    else:
        colors.append('g')
    

plt.scatter(embeddings[:, 0], embeddings[:, 1], c=colors)
plt.show()
```

```python
colors = []
for i, _ in enumerate(x_test):
    if y_train[i] == 0:
        colors.append('b')
    else:
        colors.append('g')

plt.scatter(x_test[:, 0], x_test[:, 1], c=colors)
plt.show()


embeddings = pca_model.transform(x_test)
print(embeddings.shape)

colors = []
for i, _ in enumerate(embeddings):
    if y_train[i] == 0:
        colors.append('b')
    else:
        colors.append('g')

plt.scatter(embeddings[:, 0], embeddings[:, 1], c = colors)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

model = RandomForestClassifier(n_estimators=10, max_depth=10, random_state=0, verbose=True)
model.fit(x_train, y_train)
prob_preds = model.predict_proba(x_train)
performance = roc_auc_score(y_train, prob_preds[:, 0])
print(performance)
```
